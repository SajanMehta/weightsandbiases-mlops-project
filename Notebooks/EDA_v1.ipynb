# Exploratory Data Analysis on Athlete Data Version 1

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import wandb
```

```python
# Connect to wandb
run = wandb.init(project="mlops-datasets", job_type = "load-dataset")
art = run.use_artifact("smehta15-university-of-chicago/mlops-datasets/athletes:v1")
path = art.download()
data = pd.read_csv(f"{path}/complete.csv")
run.finish()
```

```python
print(data.head())
```

```python
data.describe()
```

```python
# Count NAs
data.isnull().sum()
```

```python
print(data.nunique())
```

```python
def counter_plot(column, title=''):
    counter = column.value_counts()
    plt.figure(figsize=(8,6))
    plt.bar(counter.index, counter)
    plt.title(title)
```

```python
def point_plot(x, y):
    sns.pointplot(x = x,y = y)
    plt.show()
```

```python
counter_plot(data['gender'], 'Gender Counts')
```

```python
point_plot(data['gender'], data['total_lift'])
```

```python
plt.hist(data['total_lift'], bins = 20, range=(0,3000))
plt.title('Total lift distribution')
```

```python
plt.hist(data['height'], bins = 50, range=(0,100))
plt.title('Height Distribution (inches)')
```

```python
correlation_matrix = data.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix)
```

It appears that almost none of the features are correlated except the ones that directly add up to total lift, deadlift, squat, snatch, and clean and jerk. With these features in the dataset, I imagine a linear regression would do a fine job of "predicting" the total lift, by reducing all coefficients to 0 and holding coefficients of the internals of total lift at 1.
